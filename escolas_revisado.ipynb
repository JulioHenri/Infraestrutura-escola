#Insfraestrutura Física

```python
df1 = pd.read_excel('Escolas_REVISADO4.xlsx',sheet_name=1)
df1.head()
```

```python
df1.columns
```

```python
df1 = df1.set_index('INFRAESTRUTURA FÍSICA')
df1
```

```python
import plot_likert

myscale = ['Sim','Não']
ax = plot_likert.plot_counts(df1, myscale, plot_percentage=True);
ax.figure.set_size_inches(8, 6)
ax.xaxis.set_label_text('Porcentagem de respostas')
ax.yaxis.set_label_text('')
ax.set_yticklabels(['Energia Rede Pública','Banheiro Chuveiro','Banheiro PNE','Banheiro','Laboratório Informática','Refeitório','Despensa',
                    'Cozinha','Espaço Equipamento','Espaço Atividade','Biblioteca','Sala do Professor','Secretaria','Diretoria']);
```
