#Insfraestrutura Física

```python
df = pd.read_excel('Escolas_REVISADO4.xlsx')
df.head()
```

```python
df.info()
```

```python
df1 = df[['IN_ESPACO_ATIVIDADE', 'IN_ESPACO_EQUIPAMENTO','IN_ENERGIA_REDE_PUBLICA','IN_BANHEIRO','IN_BANHEIRO_PNE',
         'IN_BANHEIRO_CHUVEIRO', 'IN_BIBLIOTECA','IN_COZINHA', 'IN_DESPENSA','IN_LABORATORIO_INFORMATICA',
         'IN_REFEITORIO', 'IN_SALA_DIRETORIA','IN_SALA_PROFESSOR', 'IN_SECRETARIA']]

df1.head()
```

```python
df1 = df1.replace({0:'Não', 1:'Sim'})
df1
```

```python
import plot_likert

myscale = ['Não','Sim']

ax = plot_likert.plot_likert(df1, myscale, plot_percentage=True)
ax.figure.set_size_inches(8, 6) 
ax.xaxis.set_label_text('Porcentagem de respostas')
ax.yaxis.set_label_text('')
ax.set_yticklabels(['Secretaria','Sala do Professor','Diretoria','Refeitório','Laboratório Informática','Despensa','Cozinha','Biblioteca',
                    'Banheiro Chuveiro','Banheiro PNE','Banheiro','Energia Rede Pública','Espaço Equipamento','Espaço Atividade']);
```
